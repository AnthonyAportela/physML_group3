```python
import numpy as np
import pandas as pd
import os
import torch
from torch_geometric.data import Data
import networkx as nx
import subprocess
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
directory = '/home/aaportel/teams/group-3/data/'
folders = ['train_1/','train_2/','train_5/']
hit_files = sorted([f for f in os.listdir(directory + folders[0]) if f.endswith('hits.csv')])   
truth_files = sorted([f for f in os.listdir(directory + folders[0]) if f.endswith('truth.csv')])   
```

```python
X = pd.read_csv(directory + folders[0] + hit_files[0])
Y = pd.read_csv(directory + folders[0] + truth_files[0])
```

```python
def track_splitter(Y):
    unique_values = Y['particle_id'].unique()

    # Create an empty dictionary to hold the new dataframes
    new_dfs = {}

    # Loop over the unique values and create a new dataframe for each
    for val in unique_values:
        if val > 0:
        # Use boolean indexing to create a new dataframe for each value
            new_df = Y[Y['particle_id'] == val]
            # Add the new dataframe to the dictionary
            new_dfs[val] = new_df
                        
    return new_dfs
```

```python
new_dfs = track_splitter(Y)
```

```python
import random

# Create a figure and a 3D axes object
fig = plt.figure()
ax = plt.axes(projection='3d')

i=0
for df in random.sample(sorted(new_dfs), k=100):
    x = new_dfs[df]['tx']
    y = new_dfs[df]['ty']
    z = new_dfs[df]['tz']

    # Plot the data
    ax.plot3D(x, y, z)
    

# Add labels and a title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

ax.set_xlim3d(-1000,1000)
ax.set_ylim3d(-1000,1000)
ax.set_zlim3d(-2000,2000)

ax.set_title('3D plot')

# Show the plot
plt.show()
```
